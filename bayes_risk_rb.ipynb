```python
from __future__ import division, print_function
import qinfer as qi
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
try: plt.style.use('ggplot')
except: pass
```

Setup some data for a simple RB experiment.

```python
p = 0.995
A = 0.5
B = 0.5

ms = np.linspace(1, 800, 51).astype(int)
signal = A * p ** ms + B

n_shots = 25
counts = np.random.binomial(p=signal, n=n_shots)

data = np.column_stack([counts, ms, n_shots * np.ones_like(counts)])
```

Use a Bayes estimator on part of the data, and in particular extract the resulting `SMCUpdater`

```python
mean, cov, extra = qi.simple_est_rb(data, return_all=True, n_particles=12000)
print("Estimated p: {:0.4f} ± {:0.4f}".format(mean[0], np.sqrt(cov[0, 0])))
u = extra['updater']
```

Now ask for the risk of performing 20 shots of an experiment, and ask for this for each time in the list `ts`

```python
plt.plot(u.particle_weights)
risks1 = np.array(list(u.bayes_risk(np.array([(t, 20)],dtype=u.model.expparams_dtype)) for t in ms))
```

Now resample the updater (even though we have not provided any new data) and ask for the risk again.

```python
u.resample()
plt.plot(u.particle_weights)
risks2 = np.array(list(u.bayes_risk(np.array([(t, 20)],dtype=u.model.expparams_dtype)) for t in ms))
```

Now resample the updater (even though we have not provided any new data) and ask for the risk again.

```python
u.resample()
plt.plot(u.particle_weights)
risks3 = np.array(list(u.bayes_risk(np.array([(t, 20)],dtype=u.model.expparams_dtype)) for t in ms))
```

Now resample the updater (even though we have not provided any new data) and ask for the risk again.

```python
u.resample()
plt.plot(u.particle_weights)
risks4 = np.array(list(u.bayes_risk(np.array([(t, 20)],dtype=u.model.expparams_dtype)) for t in ms))
```

```python
plt.plot(ms, risks1, ms, risks2, ms, risks3, ms, risks4)
plt.legend(['1','2','3','4'])
plt.title('master branch of qinfer')
```

Now plot the result in each case. They should all be the same.

```python
plt.plot(ms, risks1, ms, risks2, ms, risks3, ms, risks4)
plt.legend(['1','2','3','4'])
plt.title('fix-bayes-risk branch of qinfer')
```
